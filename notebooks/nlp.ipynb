```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# using elbow metthod to the detrmine in a better way the number of cluster
def optimal_number_of_clusters(data, max_k):
    iters = range(2, max_k+1, 2)
    sse = []
    for k in iters:
        km = KMeans(n_clusters=k, random_state=42)
        km.fit(data)
        sse.append(km.inertia_)

    f, ax = plt.subplots(1, 1)
    ax.plot(iters, sse, marker='o')
    ax.set_xlabel('Cluster Centers')
    ax.set_xticks(iters)
    ax.set_xticklabels(iters)
    ax.set_ylabel('SSE')
    ax.set_title('SSE by Cluster Center Plot')
    plt.show()

optimal_number_of_clusters(embeddings, 10)


num_clusters = 4

# k means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(embeddings)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import umap.umap_ as umap
import plotly.express as px


# Convert the embeddings column to a numpy array format
embeddings = np.stack(df['embeddings'].apply(eval).values)

# Choose the optimal number of clusters based on the elbow method graph (e.g., 8)
num_clusters = 8

# Apply K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(embeddings)

# Function to extract keywords by cluster using TF-IDF
def get_top_keywords(data, clusters, num_clusters, n_terms):
    df_temp = pd.DataFrame(data)
    df_temp['cluster'] = clusters
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(df_temp['AbstractNarration'])
    terms = tfidf.get_feature_names_out()

    cluster_keywords = {}
    for i in range(num_clusters):
        cluster_data = tfidf_matrix[df_temp['cluster'] == i]
        tfidf_scores = np.asarray(cluster_data.sum(axis=0)).ravel()
        top_terms = tfidf_scores.argsort()[-n_terms:]
        cluster_terms = [terms[idx] for idx in top_terms]
        cluster_keywords[i] = ', '.join(cluster_terms)

    return cluster_keywords

# Extract the keywords for each cluster
cluster_keywords = get_top_keywords(df['AbstractNarration'], df['cluster'], num_clusters, 20)

# Generate automatic names for the clusters based on the most frequent keywords
def generate_cluster_names(cluster_keywords):
    cluster_names = {}
    for cluster, keywords in cluster_keywords.items():
        # Use the first three keywords as the cluster name
        keyword_list = keywords.split(', ')
        cluster_name = ' '.join(keyword_list[:3])
        cluster_names[cluster] = cluster_name.capitalize()
    return cluster_names

# Generate names for each cluster
cluster_names = generate_cluster_names(cluster_keywords)

# Assign names to the clusters
df['topic'] = df['cluster'].map(cluster_names)

# Reduce dimensionality and visualize
reducer = umap.UMAP(n_components=3, random_state=42)
embedding_3d = reducer.fit_transform(embeddings)

fig = px.scatter_3d(
    embedding_3d, x=0, y=1, z=2,
    color=df['topic'],
    labels={'color': 'Topic'}
)
fig.write_html('cluster_visualization.html')

# Save the resulting DataFrame
df.to_csv('output_with_final.csv', index=False)
```
